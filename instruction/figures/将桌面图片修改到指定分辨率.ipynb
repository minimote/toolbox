```python
import os
from PIL import Image


def change_img_size(file_path, new_width=1950, new_height=2560):
    # 打开图像文件
    image = Image.open(file_path)
    
    # 调整图像大小
    resized_image = image.resize((new_width, new_height))
    
    # 获取文件名和扩展名
    file_name, file_ext = os.path.splitext(os.path.basename(file_path))
    
    # 生成新的文件名
    new_file_name = f'new_{file_name}{file_ext}'
    new_file_path = os.path.join(os.path.dirname(file_path), new_file_name)
    
    # 保存调整后的图像
    resized_image.save(new_file_path)

def process_images_on_desktop():
    # 获取桌面路径
    desktop_path = r"E:\minimote\Desktop"
    
    # 遍历桌面路径下的所有文件
    for filename in os.listdir(desktop_path):
        # 检查是否为图片文件
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')) and filename.startswith('new_') == False:
            file_path = os.path.join(desktop_path, filename)
            # 处理图片
            change_img_size(file_path)
            print(filename)

# 调用函数处理桌面图片
process_images_on_desktop()
```
